```python
import json
import pandas as pd
from pathlib import Path
import os
os.chdir(r"C:\Users\AndreaT\Documents\football-analytics")
print("Working directory:", os.getcwd())


# Percorso base
base_path = Path("data/raw/statsbomb-open-data/data/events")

# Scegli una partita (finale Mondiali 2018)
match_id = 3879863
file_path = base_path / f"{match_id}.json"

# Apri il file JSON
with open(file_path, encoding="utf-8") as f:
    events = json.load(f)

# Controllo: quanti eventi ci sono?
print(f"Numero eventi: {len(events)}")

# Trasformo in DataFrame
df_events = pd.json_normalize(events, sep="_")

shots = df_events[df_events["type_name"] == "Shot"]
shots[["team_name", "player_name", "minute", "shot_outcome_name"]]

shots_summary = shots[[
    "team_name",
    "player_name",
    "minute",
    "second",
    "location",
    "shot_outcome_name",
    "shot_statsbomb_xg"
]]

# Visualizzo le prime righe
shots_summary.head()


```
